# SVM Classifier

# IRIS Dataset

```python
from sklearn.datasets import load_iris
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
```

```python
df=load_iris()
df1=pd.DataFrame(df.data,columns=df.feature_names)
df1['target']=df.target
X=df.data
y=df.target
print(dir(df))
sn.pairplot(df1,hue='target',palette='brg')
plt.show()
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.3)
model=SVC()
model.fit(X_train,y_train)
print('Accuracy score of Training data model is:- ',model.score(X_train,y_train))
print('Accuracy score of Testing data model is:- ',model.score(X_test,y_test))
```

```python
pred=model.predict([df.data[50],[7.7,2.6,6.9,2.3],[5.1,3.4,1.2,0.6]])
for i in pred:
    if i==0:
        print('Sample Belong to Setosa')
    elif i==1:
        print('Sample Belong to Versicolor')
    else:
        print('Sample Belong to Verginica')
```

# Digits Data

```python
from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
```

```python
df=load_digits()
print(dir(df))
print(df.data[0])
plt.matshow(df.images[0])
plt.show()
X=df.data
y=df.target
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
model=SVC()
model.fit(X_train,y_train)
print('----------------------------------------------------------------------------')
print('------------------------------------RBF----------------------------------------')
print('Accuracy score of Training data model is:- ',model.score(X_train,y_train))
print('Accuracy score of Testing data model is:- ',model.score(X_test,y_test))
print('----------------------------------------------------------------------------')
print('---------------------------------LINEAR-------------------------------------------')
model=SVC(kernel='linear')
model.fit(X_train,y_train)
print('Accuracy score of Training data model is:- ',model.score(X_train,y_train))
print('Accuracy score of Testing data model is:- ',model.score(X_test,y_test))
print('----------------------------------------------------------------------------')
print('-------------------------------------SIGMOID---------------------------------------')
model=SVC(kernel='sigmoid')
model.fit(X_train,y_train)
print('Accuracy score of Training data model is:- ',model.score(X_train,y_train))
print('Accuracy score of Testing data model is:- ',model.score(X_test,y_test))
print('----------------------------------------------------------------------------')
print('-------------------------------------POLY---------------------------------------')
model=SVC(kernel='poly')
model.fit(X_train,y_train)
print('Accuracy score of Training data model is:- ',model.score(X_train,y_train))
print('Accuracy score of Testing data model is:- ',model.score(X_test,y_test))
print('----------------------------------------------------------------------------')
print('-------------------------------------POLY With AUTO---------------------------------------')
model=SVC(kernel='poly',gamma='auto')
model.fit(X_train,y_train)
print('Accuracy score of Training data model is:- ',model.score(X_train,y_train))
print('Accuracy score of Testing data model is:- ',model.score(X_test,y_test))
```

```python
print(model.predict([X_test[100]]))
print(y_test[100])
plt.matshow(df.images[100])
plt.show()
```
