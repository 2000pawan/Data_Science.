### Difference between Loss and Error
> Error---> Residual is called Error.

> Loss----> Loss is called complete error of model.

> Loss Function/ Cost Function--->

1.OLS  2.MSE  3.RMSE  4.MAE

### Types of Loss/Error in Regression.
> Residual---------------------> Error in individual Sample. (y-pred).

> OLS(Ordinary Least Squared)--> Sum of square of residuals. (y-pred)**2

> MSE(Mean Squared Error)------> Sum of square of residuals divide by total number of sample. ((y-pred)**2)/len(y).

> RMSE(Root Mean Squared Error)-> sqrt(MSE).

> MAE(Mean Absolute Error)-------> meann(abs(residual)). (abs(y-pred)).mean().

> etc.

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error   
```

```python
df=pd.read_csv('G:/dataset/regression/salary_1_variable.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
model=LinearRegression()
model.fit(X,y)
pred=model.predict(X)
```

```python
residuals=y-pred
print(residuals)

print('---------------------------------------------------------')

ols=(residuals**2).sum()
print(ols)

print('---------------------------------------------------------')

mse=ols/len(y)
print(mse)

print('---------------------------------------------------------')

from math import sqrt
rmse=sqrt(mse)
print(rmse)

print('---------------------------------------------------------')

mae=(abs(residuals)).mean()
print(mae)

print('---------------------------------------------------------')

print(mean_absolute_error(y,pred))
print(mean_squared_error(y,pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('G:/dataset/regression/salary_1_variable.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
model=LinearRegression()
model.fit(X,y)
print(model.coef_)
print(model.intercept_)
```

# Math behind to calculate m/w/slope/intercept:-
> m= ((xmean * ymean) - xymean) / (xmeansqr-xsqrmean)

# Math behind to calculate c/b/coef :-
> c= ymean - (m * xmean)

```python
xmean=np.mean(X)
print(xmean)
ymean=np.mean(y)
print(ymean)
xymean=np.mean(X[:,0]*y)
print(xymean)
xmeansqr=(np.mean(X))**2
print(xmeansqr)
xsqrmean=np.mean((X**2))
print(xsqrmean)
```

```python
m=((xmean * ymean) - xymean) / (xmeansqr - xsqrmean)
print(m)
```

```python
c=(ymean - (m * xmean))
print(c)
```

```python
model.predict([[20]])
```

```python
result=m*20 + c
print(result)
```

### > Polynomial Features:- is used to add some feature in dataset so model train in right way. In this feature addition its use degree concept to add features like degree value is 1,2,3,4,5,6..... so if you increase degree so its add feature accoding to given degreee. commonly its used to overcome situation of underfit condition where feature too short/less.

```python
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
df=pd.read_csv('G:/dataset/regression/salary_poly.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
for i in range(1,10):
    pf=PolynomialFeatures(degree=i)
    X_new=pf.fit_transform(X)
    model=LinearRegression()
    model.fit(X_new,y)
    pred=model.predict(X_new)
    plt.scatter(X[:,0],y,c='black')
    plt.plot(X[:,0],pred,c='r')
    plt.show()
    print(r2_score(y,pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
df=pd.read_csv('G:/dataset/regression/salary_poly.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
for i in range(1,10):
    pf=PolynomialFeatures(degree=i)
    X_new_train=pf.fit_transform(X_train)
    X_new_test=pf.fit_transform(X_test)
    model=LinearRegression()
    model.fit(X_new_train,y_train)
    pred=model.predict(X_new_train)
    pred1=model.predict(X_new_test)
    print(i)
    plt.scatter(X_train[:,0],y_train,c='black')
    plt.plot(X_train[:,0],pred,c='r')
    plt.show()
    print(r2_score(y_train,pred))
    plt.scatter(X_test[:,0],y_test,c='black')
    plt.plot(X_test[:,0],pred1,c='r')
    plt.show()
    print(r2_score(y_test,pred1))
    print('..................................................................')
```

```python
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
df=pd.read_csv('G:/dataset/regression/salary_poly.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
for i in range(1,10):
    pf=PolynomialFeatures(degree=i)
    X_new_train=pf.fit_transform(X_train)
    X_new_test=pf.fit_transform(X_test)
    model=LinearRegression()
    model.fit(X_new_train,y_train)
    pred=model.predict(X_new_train)
    pred1=model.predict(X_new_test)
    print(i)
    print(r2_score(y_train,pred))
    print(r2_score(y_test,pred1))
    print('..................................................................')
```
