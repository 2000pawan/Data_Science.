```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('G:/dataset/classification/iris.csv')
df.describe().plot(kind='area',fontsize=18,figsize=(15,10),table=True,colormap='Accent')
plt.xlabel('Statistics')
plt.ylabel('Value')
plt.title('Genral Statistics of Iris DataSet')
plt.show()
```
